```python
model = AutoModelForSequenceClassification.from_pretrained(
    checkpoint, num_labels=2
)
```

```python
training_args = TrainingArguments(
    output_dir = 'training_dir',
    evaluation_strategy = 'epoch',
    save_strategy = 'epoch',
    num_train_epochs = 5,
    per_device_train_batch_size = 16,
    per_device_eval_batch_size = 64,
    logging_steps = 150 , # otherwise, 'no log' will appear under training loss
)
# the reason we need logging_step is because this dataset doesn't have a lot of samples
```

```python
from datasets import load_metric
```

```python
metric = load_metric("glue", "rte")
```

```python
# test the metric just to see what we will get back. We can see that we only see accuracy
metric.compute(predictions=[1, 0, 1], references=[1,0,0])
```

```python
from sklearn.metrics import f1_score
```

```python
def compute_metrics(logits_and_labels):
  logits, labels = logits_and_labels
  predictions = np.argmax(logits, axis=-1)
  acc = np.mean(predictions == labels)
  f1 = f1_score(labels, predictions, average='macro')
  return {'accuracy': acc, 'f1': f1}
```

```python
def tokenize_fn(batch):
  return tokenizer(batch['sentence1'], batch['sentence2'], truncation=True)
```

```python
tokenized_datasets = raw_datasets.map(tokenize_fn, batched=True)
```

```python
trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation'],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
